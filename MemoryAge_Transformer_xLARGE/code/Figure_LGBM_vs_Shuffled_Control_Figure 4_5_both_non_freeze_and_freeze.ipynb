```python
plt.figure(figsize=(10, 8))

# ---------- non-freeze ----------
non_freeze_shuffled_control = np.array([0.481, 0.408, 0.494, 0.35, 0.486, 0.497, 0.431, 0.417, 0.494, 0.493, 0.509, 0.51])
non_freeze_trans = np.array([0.88, 0.811, 0.904, 0.787, 0.825, 0.83, 0.869, 0.813, 0.61, 0.683, 0.572, 0.783])

non_freeze_trans_mean = np.mean(non_freeze_trans)
non_freeze_trans_std = np.std(non_freeze_trans)
non_freeze_shuffled_control_mean = np.mean(non_freeze_shuffled_control)
non_freeze_shuffled_control_std = np.std(non_freeze_shuffled_control)

plt.bar([1,2], [non_freeze_trans_mean, non_freeze_shuffled_control_mean], yerr=[non_freeze_trans_std, non_freeze_shuffled_control_std], width=0.5, color=['chocolate','chocolate'])

# make a cap for the error bar
plt.errorbar([1,2], [non_freeze_trans_mean, non_freeze_shuffled_control_mean], yerr=[non_freeze_trans_std, non_freeze_shuffled_control_std], fmt='none', ecolor='black', capsize=7, capthick=2)

# # plot the scatter plot of each group
# plt.scatter(np.ones(len(non_freeze_shuffled_control))*2, non_freeze_shuffled_control, color='black', marker='o')
# plt.scatter(np.ones(len(non_freeze_trans)), non_freeze_trans, color='black', marker='o')

# connect each pair of points from non_freeze_shuffled_control and non_freeze_trans by a dash line
for i in range(len(non_freeze_shuffled_control)):
    plt.plot([1,2], [non_freeze_trans[i], non_freeze_shuffled_control[i]], color='grey', linestyle='--')

# do a wilcoxon signed rank sum test of the groups, and print the p-value
stat, p = wilcoxon(non_freeze_trans, non_freeze_shuffled_control)
print('----- non-freeze -----')
print('stat: ', stat)
print('p-value: ', p) 

# ---------- freeze --------------
freeze_shuffled_control = np.array([0.369, 0.508, 0.439, 0.351, 0.356, 0.372, 0.482, 0.385, 0.359, 0.415, 0.366, 0.359])
freeze_trans = np.array([0.787, 0.85, 0.743, 0.544, 0.609, 0.685, 0.664, 0.516, 0.505, 0.485, 0.662, 0.682])

freeze_trans_mean = np.mean(freeze_trans)
freeze_trans_std = np.std(freeze_trans)
freeze_shuffled_control_mean = np.mean(freeze_shuffled_control)
freeze_shuffled_control_std = np.std(freeze_shuffled_control)

plt.bar([3,4], [freeze_trans_mean, freeze_shuffled_control_mean], yerr=[freeze_trans_std, freeze_shuffled_control_std], width=0.5, color=['darkviolet','darkviolet'])

# make a cap for the error bar
plt.errorbar([3,4], [freeze_trans_mean, freeze_shuffled_control_mean], yerr=[freeze_trans_std, freeze_shuffled_control_std], fmt='none', ecolor='black', capsize=7, capthick=2)

plt.xticks([1,2,3,4],['Correct\nLabel', 'Label\nshuffled','Correct\nLabel', 'Label\nshuffled'], font='Arial', fontsize=28)

# # plot the scatter plot of each group
# plt.scatter(np.ones(len(freeze_shuffled_control))*4, freeze_shuffled_control, color='black', marker='o')
# plt.scatter(np.ones(len(freeze_trans))*3, freeze_trans, color='black', marker='o')

# connect each pair of points from freeze_shuffled_control and freeze_trans by a dash line
for i in range(len(freeze_shuffled_control)):
    plt.plot([3,4], [freeze_trans[i], freeze_shuffled_control[i]], color='grey', linestyle='--')

# do a wilcoxon signed rank sum test of the groups, and print the p-value
stat, p = wilcoxon(freeze_trans, freeze_shuffled_control)
print('----- freeze -----')
print('stat: ', stat)
print('p-value: ', p)

# --------------------------------

plt.ylabel('F1 score', font='Arial', fontsize=28)
# plt.title('LightGBM', font='Arial', fontsize=30)
# make the bars' width smaller
plt.xlim(0.5,4.5)
plt.ylim(0,1.1)

# make the y axis font size larger
plt.yticks(font='Arial', fontsize=28)

plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)


plt.show()
```
